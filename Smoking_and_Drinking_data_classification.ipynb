```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Smoking and Drinking.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl =  LabelEncoder()
```

```python
df['sex'] = lbl.fit_transform(df['sex'])
df['DRK_YN'] = lbl.fit_transform(df['DRK_YN'])
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan , strategy = 'mean')
df.iloc[:,:] = si.fit_transform(df.iloc[:,:])
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize = (15,15))
pl_num = 1
for i in df.columns:
  if pl_num <= 22:
    plt.subplot(5,5,pl_num)
    sns.lineplot(df[i])
  pl_num +=1
plt.show()
```

```python
plt.figure(figsize = (15,15))
pl_num = 1
for i in df.columns:
  if pl_num <= 22:
    plt.subplot(5,5,pl_num)
    sns.boxplot(df[i])
  pl_num +=1
plt.show()
```

```python
plt.figure(figsize=(15,15))
crr = df.corr()
sns.heatmap(crr,annot = True)
```

```python
X = df.drop('DRK_YN',axis =1)
Y = df['DRK_YN']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(X,Y,test_size = .2)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'log_loss')
dt.fit(x_train,y_train)
```

```python
dt_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(dt_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=145,criterion = 'entropy',max_features = 10)
rf.fit(x_train,y_train)
```

```python
rf_pre = rf.predict(x_test)
```

```python
accuracy_score(rf_pre,y_test)
```
